```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

# transform features function
def transform_features(df):
# select the numerical columns
    null_vals_numeric = df.select_dtypes(include = ['float64','int64']).isnull().sum()
##null_vals = null_vals[null_vals > 2] # this function will segregate the values and print only the values that are true
    # percentage of null values in the numerical column 
    prcnt_null = ((null_vals_numeric)*100/len(housing)) ## calculating the percent of null values
    
    #Replace the null values with the mode of the column if it has less than 5% of null values 
    null_5_cols = prcnt_null[(prcnt_null < 5) & (prcnt_null>0)].index.tolist()
    for col in null_5_cols:
        df[col] = df[col].fillna(df[col].mode()[0])
        
        #Drop the columns if they have more than 25 percent of null values
    null_vals_p = df.isnull().sum()*100/len(housing)
    nul_25_cols = null_vals_p[null_vals_p>=25].index.tolist()
    df = df.drop(columns = nul_25_cols)
# drop the rows with the remaining null values 
    df = df.dropna()
     
    return df


def select_features(df):
    numeric_vals = df.select_dtypes(include = ['float64','int64'])
    
    corr_sale_price = np.abs(numeric_vals.corr()[["SalePrice"]])
    cols_to_keep = corr_sale_price[corr_sale_price > 0.4].dropna().index.tolist()
    
    return df[cols_to_keep]
## p is the parameter passed in the function 
def train_and_test(data, p=0):
    
    if p == 0:
    
        X_train = data.drop(columns = ["SalePrice"]).iloc[:1460 , :] # Assigning first 1460 rows in training set
        X_test = data.drop(columns = ['SalePrice']).iloc[1460: , :] # Assigning the rows after 1460 in test set

        Y_train = data["SalePrice"].iloc[:1460]
        Y_test = data['SalePrice'].iloc[1460:]

        lr=LinearRegression()

        lr.fit(X_train , Y_train)
        Y_pred = lr.predict(X_test)
        score = r2_score(Y_test, Y_pred)
        
    elif p==1:
        #Shuffle the dataset
        data = data.sample(frac=1).reset_index(drop=True)
        
        #create the train set
        fold_one = data.drop(columns = ["SalePrice"]).iloc[:1460 , :] # Assigning first 1460 rows in training set
        fold_two= data.drop(columns = ['SalePrice']).iloc[1460: , :] # Assigning the rows after 1460 in test set
        # create the test set
        y_one = data["SalePrice"].iloc[:1460]
        Y_two = data['SalePrice'].iloc[1460:]
        # train the linear regression model
        lr=LinearRegression()
        
        #Calculate the root mean squared error
        lr.fit(fold_one , y_one)
        Y_pred = lr.predict(fold_two)
        score = r2_score(Y_two, Y_pred)
       
    else:
        lr = LinearRegression()
        scores = cross_val_score(lr , data.drop(columns=["SalePrice"]), data["SalePrice"],scoring="r2",cv=p) # training on all columns except the SalePrice and predicting on Sale Price
        score = np.mean(scores)
    return score
```

```python
housing_clean = transform_features(housing)
```

```python
#Filter the dataset to keep on;y the above columns
housing_clean_f1 = select_features(housing_clean)
housing_clean_f1
```

```python
score_1 = train_and_test(housing_clean_f1)
print(score_1)
```

```python
score_2 = train_and_test(housing_clean_f1, p=1)
print(score_2)
```
